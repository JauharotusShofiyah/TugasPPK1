### TUGAS MATPLOTLIB

##### 1. Cari bagaimana cara menampilkan Judul (title) dari grafik

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5,6,7])
y = np.array([10,20,30,40,50,60,70])

plt.title("PENJUALAN KAOS SELAMA 1 MINGGU")

plt.plot(x,y)
plt.show()
```

##### 2. Cari bagaimana cara memberikan label pada sumbu x dan sumbu y

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5,6,7])
y = np.array([10,20,30,40,50,60,70])

plt.title("PENJUALAN KAOS SELAMA 1 MINGGU")
plt.xlabel("Tanggal")
plt.ylabel("Jumlah Kaos Yang Terjual")

plt.plot(x,y)
plt.show()
```

##### 3. Cari bagaimana menambahkan teks dalam grafik

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5,6,7])
y = np.array([20,40,10,50,30,60,70])

plt.title("PENJUALAN KAOS SELAMA 1 MINGGU")
plt.xlabel("Tanggal")
plt.ylabel("Jumlah Kaos Yang Terjual")
plt.text(1,20,"20")
plt.text(2,40,"40")
plt.text(3,10,"10")
plt.text(4,50,"50")
plt.text(5,30,"30")
plt.text(6,60,"60")
plt.text(7,70,"70")

plt.plot(x,y)
plt.show()
```
